```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from statistics import mean, stdev
from process_rolls import ROLLS_FILE, MIN_DIE, MAX_DIE, SUCCESS_THRESHOLD, USERS, count_successes, line_contains_user, is_dice_roll_results, parse_dice_roll_result, user_from_line
```

```python
with open(ROLLS_FILE) as f:
    data = f.readlines()
```

```python
individual_rolls_by_user = {user: [] for user in USERS}
for line in data:
    if line_contains_user(line):
        user = user_from_line(line)
    elif is_dice_roll_results(line):
        individual_rolls_by_user[user].append(parse_dice_roll_result(line))
```

```python
successes_per_roll_per_user = {user: [] for user in USERS}
for user, list_of_rolls in individual_rolls_by_user.items():
    for roll_set in list_of_rolls:
        successes_per_roll_per_user[user].append(count_successes(roll_set, calc_percent=False))
```

```python
users = []
avg_successes = []
stdev_successes = []
for user, num_successes in successes_per_roll_per_user.items():
    users.append(user)
    avg_successes.append(mean(num_successes))
    stdev_successes.append(stdev(num_successes))
```

```python
sns.set_context("talk")
plt.figure(figsize=(12,12))
```

```python
plt.barh(
    y=range(len(users)),
    width=avg_successes,
    xerr=stdev_successes
)
plt.yticks(range(len(users)), [users[i] for i in range(len(users))])
plt.xlabel("Average (+/- std. dev.) successes per roll")
```

```python
for user in users:
    sns.distplot(successes_per_roll_per_user[user])
    plt.title(user)
    plt.show()
```

```python
percent_successes_per_roll_per_user = {user: [] for user in USERS}
for user, list_of_rolls in individual_rolls_by_user.items():
    for roll_set in list_of_rolls:
        percent_successes_per_roll_per_user[user].append(count_successes(roll_set))
```

```python
users = []
avg_successes = []
stdev_successes = []
for user, num_successes in percent_successes_per_roll_per_user.items():
    users.append(user)
    avg_successes.append(mean(num_successes))
    stdev_successes.append(stdev(num_successes))
```

```python
plt.barh(
    y=range(len(users)),
    width=avg_successes,
    xerr=stdev_successes
)
plt.yticks(range(len(users)), [users[i] for i in range(len(users))])
plt.xlabel("Average (+/- std. dev.) % successes per roll")
```
